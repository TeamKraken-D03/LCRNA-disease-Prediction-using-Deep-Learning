```python
import pandas as pd
from sqlalchemy import create_engine
import mysql.connector
import pymysql
import urllib.parse

# Step 1: Load the dataset
tsv_file_path = "website_alldata.tsv"  # Replace with your TSV file path
df = pd.read_csv(tsv_file_path, delimiter="\t")

# Step 2: Define the MySQL connection configuration
password = urllib.parse.quote("Naren@4010")  # URL-encode the password

config = {
    "user": "Naren",
    "password": password,
    "host": "127.0.0.1",
    "database": "ncRNA_Database",
    "port": 3306,
}

try:
    # Step 3: Create a MySQL connection
    connection = mysql.connector.connect(
        user=config["user"],
        password="Naren@4010",  # Use decoded password here
        host=config["host"],
        database=config["database"],
        port=config["port"],
    )

    if connection.is_connected():
        print("Connected to MySQL successfully!")

    # Step 4: Create SQLAlchemy engine
    connection_string = f"mysql+pymysql://{config['user']}:{config['password']}@{config['host']}:{config['port']}/{config['database']}"
    engine = create_engine(connection_string)

    # Step 5: Push data in chunks to prevent overload
    table_name = "ncRNA_Data"
    df.to_sql(table_name, con=engine, if_exists="append", index=False, chunksize=1000)

    print("Data successfully pushed to MySQL!")

except mysql.connector.Error as err:
    print(f"MySQL Connection Error: {err}")

except Exception as e:
    print(f"Unexpected Error: {e}")

finally:
    # Step 6: Close the connection
    if "connection" in locals() and connection.is_connected():
        connection.close()
        print("MySQL connection closed.")
```
